```python
import json
import requests
from pprint import pprint
import pandas as pd
```

```python
base_url="http://crashviewer.nhtsa.dot.gov/CrashAPI/crashes/GetCaseList?"

params = {
    'states':'36',
    'fromYear':'',
    'toYear':'',
    'minNumOfVehicles':'1',
    'maxNumOfVehicles':'10',
    'format':'json'
}

response = requests.get(base_url,params=params)

years = ['2014','2015','2016','2017','2018']

county=[]
CrashDate=[]
Fatals=[]
TotalVehicles=[]

for year in years:
    params['fromYear'] = year
    params['toYear'] = year
    
    response = requests.get(base_url,params=params).json()
   
    for county_name in response['Results'][0]:
        county.append(county_name["CountyName"])
        CrashDate.append(county_name["CrashDate"])
        Fatals.append(county_name["Fatals"])
        TotalVehicles.append(county_name["TotalVehicles"])
```

```python
county
```

```python
NY_dict = {
    "county_name": county,
    "Crash_date":CrashDate ,
    "Fatals": Fatals,
    "Total_vehicles":TotalVehicles
}
NY_data =pd.DataFrame(NY_dict)
NY_data
```

```python
NY_county=NY_data.loc[(NY_data["county_name"]=="RICHMOND (85)") | (NY_data["county_name"]=="NEW YORK (61)") |\
              (NY_data["county_name"]=="BRONX (5)") | (NY_data["county_name"]=="KINGS (47)") |\
              (NY_data["county_name"]=="QUEENS (81)") ,:]
NY_county                                     
```
